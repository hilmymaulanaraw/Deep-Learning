```python
import os

def clean_and_prepare(df, base_path):
    # Combine labels: treating BENIGN and BENIGN_WITHOUT_CALLBACK as 0, MALIGNANT as 1
    df['label'] = df['pathology'].apply(lambda x: 1 if 'MALIGNANT' in x else 0)

    # Function to find actual file path since the CSV path might be slightly different from folder structure
    def get_full_path(row):
        # In this dataset, the 'image file path' column typically refers to the parent folder name in 'jpeg/'
        folder_part = row['image file path'].split('/')[0]
        search_dir = os.path.join(base_path, 'jpeg', folder_part)
        if os.path.exists(search_dir):
            for r, d, f in os.walk(search_dir):
                for file in f:
                    if file.endswith('.jpg'):
                        return os.path.join(r, file)
        return None

    df['full_path'] = df.apply(get_full_path, axis=1)
    return df.dropna(subset=['full_path'])

# Combine and clean all datasets
train_df = pd.concat([calc_train, mass_train], ignore_index=True)
test_df = pd.concat([calc_test, mass_test], ignore_index=True)

train_df_cleaned = clean_and_prepare(train_df, path)
test_df_cleaned = clean_and_prepare(test_df, path)

print(f"Training samples after cleaning: {len(train_df_cleaned)}")
print(f"Testing samples after cleaning: {len(test_df_cleaned)}")
print("\nClass Distribution in Train:", train_df_cleaned['label'].value_counts())
print(train_df_cleaned[['full_path', 'label']].head())
```

```python
import os

def fix_path_mapping(df, base_path):
    # Extract the second part of the path (the UID folder) from 'image file path'
    # e.g., 'Calc-Training_P_00005_RIGHT_CC/1.3.6.1.../1.3.6.1.../000000.dcm'
    def get_uid(p):
        if isinstance(p, str):
            parts = p.split('/')
            if len(parts) > 1: return parts[1]
        return None

    df['uid'] = df['image file path'].apply(get_uid)

    # Iterate through jpeg directory to map UID folders to actual file paths
    jpeg_dir = os.path.join(base_path, 'jpeg')
    uid_to_file = {}
    for root, dirs, files in os.walk(jpeg_dir):
        for f in files:
            if f.endswith('.jpg'):
                # The folder name containing the file or its parent is the UID
                parent_folder = os.path.basename(root)
                uid_to_file[parent_folder] = os.path.join(root, f)

    df['full_path'] = df['uid'].map(uid_to_file)
    df['label'] = df['pathology'].apply(lambda x: 1 if 'MALIGNANT' in str(x) else 0)
    return df.dropna(subset=['full_path']).copy()

train_df_cleaned = fix_path_mapping(train_df, path)
test_df_cleaned = fix_path_mapping(test_df, path)

print(f"Success! Train samples: {len(train_df_cleaned)}")
print(f"Success! Test samples: {len(test_df_cleaned)}")
if len(train_df_cleaned) > 0:
    print("\nExample paths:")
    print(train_df_cleaned[['full_path', 'label']].head())
    print("\nClass Distribution:")
    print(train_df_cleaned['label'].value_counts())
```

```python
import os
import pandas as pd

def map_images_to_metadata(df, base_path):
    # Extract all possible UIDs from the 'image file path' column
    # The format is 'Project_Patient_Laterality_View/UID1/UID2/000000.dcm'
    def get_potential_uids(p):
        if isinstance(p, str):
            return set(p.split('/'))
        return set()

    # Map every UID folder in the jpeg directory to its actual .jpg file path
    jpeg_dir = os.path.join(base_path, 'jpeg')
    uid_map = {}
    for root, dirs, files in os.walk(jpeg_dir):
        for f in files:
            if f.endswith('.jpg'):
                # Use the immediate parent folder name as the key (UID)
                uid_map[os.path.basename(root)] = os.path.join(root, f)

    def match_path(p):
        uids = get_potential_uids(p)
        for uid in uids:
            if uid in uid_map:
                return uid_map[uid]
        return None

    df = df.copy()
    df['full_path'] = df['image file path'].apply(match_path)
    df['label'] = df['pathology'].apply(lambda x: 1 if 'MALIGNANT' in str(x) else 0)

    cleaned_df = df.dropna(subset=['full_path']).copy()
    return cleaned_df

# Re-process train and test data
train_df_cleaned = map_images_to_metadata(train_df, path)
test_df_cleaned = map_images_to_metadata(test_df, path)

print(f"Finalized Train samples: {len(train_df_cleaned)}")
print(f"Finalized Test samples: {len(test_df_cleaned)}")

if len(train_df_cleaned) > 0:
    print("\nSample distribution:")
    print(train_df_cleaned['label'].value_counts())
    print("\nFirst few mapped entries:")
    print(train_df_cleaned[['full_path', 'label']].head())
else:
    print("\nMapping still failed. Listing first 10 keys in uid_map for debugging:")
    # Recalculate map locally to show debug info
    debug_map = {}
    for root, dirs, files in os.walk(os.path.join(path, 'jpeg')):
        if files: debug_map[os.path.basename(root)] = files[0]
        if len(debug_map) > 10: break
    print(list(debug_map.keys()))
```

### 5. Preprocessing Gambar dan Pembuatan Dataset TensorFlow (ROI & CLAHE)

```python
import tensorflow as tf
import cv2
import numpy as np
import pandas as pd
import os

# 1. Re-load the metadata CSVs to restore kernel state
csv_dir = os.path.join(path, 'csv')
calc_train = pd.read_csv(os.path.join(csv_dir, 'calc_case_description_train_set.csv'))
calc_test = pd.read_csv(os.path.join(csv_dir, 'calc_case_description_test_set.csv'))
mass_train = pd.read_csv(os.path.join(csv_dir, 'mass_case_description_train_set.csv'))
mass_test = pd.read_csv(os.path.join(csv_dir, 'mass_case_description_test_set.csv'))

# 2. Re-establish cleaned dataframes
train_df = pd.concat([calc_train, mass_train], ignore_index=True)
train_df_cleaned = map_images_to_metadata(train_df, path)
test_df = pd.concat([calc_test, mass_test], ignore_index=True)
test_df_cleaned = map_images_to_metadata(test_df, path)

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

def preprocess_pipeline(image):
    # Nested function to ensure it's in scope for tf.py_function
    def remove_artifacts_and_extract_roi_local(image_arg):
        # 1. Thresholding to create a mask of the breast
        gray = cv2.cvtColor(np.uint8(image_arg * 255), cv2.COLOR_RGB2GRAY)
        _, thresh = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)

        # 2. Find the largest contour (the breast)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if not contours:
            return image_arg

        largest_cnt = max(contours, key=cv2.contourArea)
        mask = np.zeros_like(gray)
        cv2.drawContours(mask, [largest_cnt], -1, 255, -1)

        # 3. Apply mask
        result = cv2.bitwise_and(np.uint8(image_arg * 255), np.uint8(image_arg * 255), mask=mask)

        # 4. Crop to bounding box
        x, y, w, h = cv2.boundingRect(largest_cnt)
        roi = result[y:y+h, x:x+w]

        # Resize back to target size
        roi_resized = cv2.resize(roi, (224, 224))
        return np.float32(roi_resized) / 255.0

    image_roi = remove_artifacts_and_extract_roi_local(image)
    img_uint8 = np.uint8(image_roi * 255)
    denoised = cv2.medianBlur(img_uint8, 3)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    channels = cv2.split(denoised)
    clahe_channels = [clahe.apply(c) for c in channels]
    result = cv2.merge(clahe_channels)
    return np.float32(result) / 255.0

def load_and_preprocess_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = img / 255.0
    [img,] = tf.py_function(preprocess_pipeline, [img], [tf.float32])
    img.set_shape((*IMG_SIZE, 3))
    return img, label

def create_dataset(df, shuffle=True):
    paths = df['full_path'].values
    labels = df['label'].values
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle: ds = ds.shuffle(buffer_size=1000)
    ds = ds.batch(BATCH_SIZE).prefetch(buffer_size=tf.data.AUTOTUNE)
    return ds

if len(train_df_cleaned) > 0:
    train_ds = create_dataset(train_df_cleaned, shuffle=True)
    test_ds = create_dataset(test_df_cleaned, shuffle=False)
    print(f"Pipeline sukses. Sampel latih: {len(train_df_cleaned)}")
else:
    print("Error: train_df_cleaned masih kosong. Cek logika pemetaan.")
```

### 7. Membangun dan Melatih Model CNN DenseNet121

```python
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model
from tensorflow.keras import layers, optimizers

# Using DenseNet121 for better feature propagation in medical imaging
base_model = DenseNet121(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False

x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.4)(x) # Increased dropout for regularization
predictions = layers.Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# Learning rate scheduler can be added here for optimization
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc'), tf.keras.metrics.Recall(name='recall'), tf.keras.metrics.Precision(name='precision')]
)

print("Model updated to DenseNet121 with clinical metrics.")
epochs = 15
history = model.fit(augmented_train_ds, validation_data=test_ds, epochs=epochs)
```

### 10. Implementasi Arsitektur Attention U-Net

```python
from tensorflow.keras import layers, models, backend as K
import tensorflow as tf

def attention_block(x, gating, inter_shape):
    shape_x = K.int_shape(x)
    phi_g = layers.Conv2D(inter_shape, (1, 1), padding='same')(gating)
    theta_x = layers.Conv2D(inter_shape, (1, 1), strides=(1, 1), padding='same')(x)
    concat_xg = layers.add([theta_x, phi_g])
    act_xg = layers.Activation('relu')(concat_xg)
    psi_xg = layers.Conv2D(1, (1, 1), padding='same')(act_xg)
    sigmoid_xg = layers.Activation('sigmoid')(psi_xg)
    y = layers.multiply([sigmoid_xg, x])
    result = layers.Conv2D(shape_x[3], (1, 1), padding='same')(y)
    result_bn = layers.BatchNormalization()(result)
    return result_bn

def conv_block(x, filters):
    x = layers.Conv2D(filters, (3, 3), padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Conv2D(filters, (3, 3), padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    return x

def build_attention_unet_classifier(input_shape=(224, 224, 3)):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    c2 = conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    c3 = conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    c4 = conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bridge
    b1 = conv_block(p4, 1024)

    # Decoder
    u1 = layers.Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(b1)
    att1 = attention_block(x=c4, gating=u1, inter_shape=512)
    c5 = conv_block(layers.concatenate([u1, att1]), 512)

    u2 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c5)
    att2 = attention_block(x=c3, gating=u2, inter_shape=256)
    c6 = conv_block(layers.concatenate([u2, att2]), 256)

    u3 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c6)
    att3 = attention_block(x=c2, gating=u3, inter_shape=128)
    c7 = conv_block(layers.concatenate([u3, att3]), 128)

    u4 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c7)
    att4 = attention_block(x=c1, gating=u4, inter_shape=64)
    c8 = conv_block(layers.concatenate([u4, att4]), 64)

    # Classifier Head (Added to match scalar labels)
    # We keep a named conv layer for Grad-CAM access
    last_conv = layers.Conv2D(64, (3, 3), padding='same', activation='relu', name='last_conv_layer')(c8)
    gap = layers.GlobalAveragePooling2D()(last_conv)
    output = layers.Dense(1, activation='sigmoid', name='final_output')(gap)

    model = models.Model(inputs=[inputs], outputs=[output], name="Attention_UNet_Classifier")
    return model

model_attn = build_attention_unet_classifier()
print("Attention U-Net Classifier successfully updated to handle scalar labels.")
```

```python
import cv2
import numpy as np

def remove_artifacts_and_extract_roi(image):
    # 1. Thresholding to create a mask of the breast
    gray = cv2.cvtColor(np.uint8(image * 255), cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)

    # 2. Find the largest contour (the breast)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return image

    largest_cnt = max(contours, key=cv2.contourArea)
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [largest_cnt], -1, 255, -1)

    # 3. Apply mask
    result = cv2.bitwise_and(np.uint8(image * 255), np.uint8(image * 255), mask=mask)

    # 4. Crop to bounding box
    x, y, w, h = cv2.boundingRect(largest_cnt)
    roi = result[y:y+h, x:x+w]

    # Resize back to target size
    roi_resized = cv2.resize(roi, (224, 224))
    return np.float32(roi_resized) / 255.0

print("Advanced Preprocessing (ROI & Artifact Removal) functions defined.")
```

```python
import tensorflow as tf
from tensorflow.keras import backend as K

def dice_loss(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    numerator = 2 * tf.reduce_sum(y_true * y_pred)
    denominator = tf.reduce_sum(y_true + y_pred)
    return 1 - (numerator + 1e-6) / (denominator + 1e-6)

# Re-compile Attention U-Net with Dice Loss
model_attn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss=dice_loss,
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

print("Attention U-Net re-compiled with Dice Loss for medical segmentation optimization.")
```

```python
from sklearn.metrics import precision_recall_curve

def tune_threshold(model, dataset):
    y_true = []
    y_scores = []

    for imgs, labels in dataset:
        preds = model.predict(imgs, verbose=0)
        y_true.extend(labels.numpy())
        y_scores.extend(preds.flatten())

    precisions, recalls, thresholds = precision_recall_curve(y_true, y_scores)
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-10)
    best_idx = np.argmax(f1_scores)
    best_threshold = thresholds[best_idx]

    print(f"Optimal Threshold found: {best_threshold:.4f}")
    print(f"Best F1-Score: {f1_scores[best_idx]:.4f}")
    return best_threshold

# threshold_opt = tune_threshold(model_attn, test_ds)
```

### 11. Interpretabilitas Model dengan Grad-CAM

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import tensorflow as tf

def get_gradcam(img_array, model, last_conv_layer_name=None):
    # Jika nama layer tidak diberikan, cari layer konvolusi terakhir secara otomatis
    if last_conv_layer_name is None:
        for layer in reversed(model.layers):
            if isinstance(layer, tf.keras.layers.Conv2D):
                last_conv_layer_name = layer.name
                break

    print(f"Menggunakan layer untuk Grad-CAM: {last_conv_layer_name}")

    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = tf.reduce_mean(predictions)

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-10)
    return heatmap.numpy()

# Ambil contoh gambar dari test_ds
for images, labels in test_ds.take(1):
    sample_img = images[0:1]
    sample_label = labels[0]

# Memanggil Grad-CAM (biarkan mencari layer otomatis)
heatmap = get_gradcam(sample_img, model_attn)

# Visualisasi
img_np = sample_img[0].numpy()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(img_np)
plt.title(f"Original Image (Label: {'Malignant' if sample_label==1 else 'Benign'})")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_np)
# Resize heatmap agar sesuai ukuran gambar input (224, 224)
resized_heatmap = tf.image.resize(heatmap[..., np.newaxis], (224, 224)).numpy()
plt.imshow(resized_heatmap, alpha=0.5, cmap='jet')
plt.title("Attention U-Net Grad-CAM")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def get_gradcam(img_array, model, last_conv_layer_name=None):
    if last_conv_layer_name is None:
        for layer in reversed(model.layers):
            if isinstance(layer, tf.keras.layers.Conv2D):
                last_conv_layer_name = layer.name
                break

    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = tf.reduce_mean(predictions)

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-10)
    return heatmap.numpy()

def visualize_attention_unet_internals(model, dataset):
    for images, labels in dataset.take(1):
        img = images[0:1]
        label = labels[0]
        break

    layer_outputs = []
    layer_names = []

    for layer in model.layers:
        if len(layer.output.shape) == 4:
            if 'conv2d_7' in layer.name or 'last_conv_layer' in layer.name:
                layer_outputs.append(layer.output)
                layer_names.append(layer.name)
            if 'activation' in layer.name and 'sigmoid' in layer.name and 'final' not in layer.name:
                layer_outputs.append(layer.output)
                layer_names.append("Attention Gate")

    if not layer_outputs:
        print("Tidak ditemukan layer yang cocok untuk visualisasi.")
        return

    vis_model = tf.keras.models.Model(inputs=model.input, outputs=layer_outputs)
    activations = vis_model.predict(img)
    if not isinstance(activations, list):
        activations = [activations]

    plt.figure(figsize=(24, 6))

    plt.subplot(1, 4, 1)
    plt.imshow(img[0].numpy())
    plt.title(f"Original Image\nLabel: {'Malignant' if label==1 else 'Benign'}")
    plt.axis('off')

    plt.subplot(1, 4, 2)
    feat_map = activations[0][0]
    if len(feat_map.shape) == 3:
        feat_map = np.mean(feat_map, axis=-1)
    plt.imshow(feat_map, cmap='viridis')
    plt.title(f"Feature Map: {layer_names[0]}")
    plt.axis('off')

    plt.subplot(1, 4, 3)
    attn_map = activations[-1][0]
    if len(attn_map.shape) == 3:
        attn_map = np.mean(attn_map, axis=-1)
    plt.imshow(attn_map, cmap='hot')
    plt.title("Attention Gate Map")
    plt.axis('off')

    plt.subplot(1, 4, 4)
    heatmap = get_gradcam(img, model)
    resized_heatmap = tf.image.resize(heatmap[..., np.newaxis], (224, 224)).numpy()
    plt.imshow(img[0].numpy())
    plt.imshow(resized_heatmap, alpha=0.5, cmap='jet')
    plt.title("Grad-CAM Focus")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

visualize_attention_unet_internals(model_attn, test_ds)
```

### 14. Membangun dan Melatih Model CNN ResNet50

ResNet (Residual Networks) adalah arsitektur CNN lain yang sangat efektif, terutama untuk jaringan yang sangat dalam. Kunci dari ResNet adalah penggunaan 'residual connections' atau 'skip connections' yang memungkinkan informasi melewati beberapa lapisan. Ini membantu mengatasi masalah vanishing gradient dan memungkinkan pelatihan model yang lebih dalam dan lebih akurat. Kami akan menggunakan ResNet50 yang sudah dilatih sebelumnya dari `imagenet` sebagai feature extractor.

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras import layers, optimizers

# Memuat ResNet50 sebagai model dasar
# input_shape harus sesuai dengan ukuran gambar Anda, include_top=False untuk tidak menyertakan lapisan klasifikasi ImageNet
# weights='imagenet' untuk menggunakan bobot yang sudah dilatih sebelumnya
base_model_resnet = ResNet50(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Bekukan lapisan base_model agar tidak dilatih ulang
base_model_resnet.trainable = False

# Tambahkan lapisan klasifikasi kustom di atas output ResNet50
x = base_model_resnet.output
x = layers.GlobalAveragePooling2D()(x) # Mengurangi dimensi spasial menjadi vektor fitur
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x) # Lapisan Dense dengan 256 unit dan aktivasi ReLU
x = layers.Dropout(0.4)(x) # Dropout untuk regularisasi
predictions = layers.Dense(1, activation='sigmoid')(x) # Lapisan output untuk klasifikasi biner dengan aktivasi Sigmoid

# Buat model lengkap
model_resnet = Model(inputs=base_model_resnet.input, outputs=predictions, name="ResNet50_Classifier")

# Kompilasi model
model_resnet.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc'), tf.keras.metrics.Recall(name='recall'), tf.keras.metrics.Precision(name='precision')]
)

print("Model ResNet50 berhasil dibuat dan dikompilasi.")

# Latih model
epochs_resnet = 15 # Sesuaikan jumlah epoch jika diperlukan
history_resnet = model_resnet.fit(augmented_train_ds, validation_data=test_ds, epochs=epochs_resnet)
```
